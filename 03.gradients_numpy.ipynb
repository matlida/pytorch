```python
# 使用numpy进行梯度下降
import numpy as np
# set  f = 2 * x 
# training 
x = np.array([1,2,3,4],dtype = np.float32)
y = np.array([2,4,6,8],dtype = np.float32)

#初始化权重
w = 0.00

# 向前传递
def forward(x):
    return w*x

# 计算损失函数
def loss(y,y_hat):
    return ((y_hat - y)**2).mean()

# mse = 1/n *(w*x - y)
# dJ/dw = 

def grad(x,y,y_hat):
    return np.dot(2*x, y_hat-y).mean()

print(f'训练前的预测值： f(5) = {forward(5):.3f}')

learning_rate = 0.05
n_iters = 20

for epoch in range(n_iters):
    y_pred = forward(x)
    # loss
    l = loss(x,y_pred)
    # grad
    dw = grad(x,y,y_pred)
    # update weights
    w -= learning_rate*dw
    
    if epoch%1==0:
        print(f'epoch {epoch+1}:w = {w:.3f},loss = {l:.8f}')

print(f'训练后的预测值为：f(5) =  {forward(5):.3f}')



# 这里是有问题的，应该随着的迭代次数增加，loss是越来越小的,应该是梯度的计算的定义有些问题
```

```python
# 使用torch进行自动的梯度下降
import torch
# set  f = 2 * x 
# training 
x = torch.tensor([1,2,3,4],dtype = torch.float32)
y = torch.tensor([2,4,6,8],dtype = torch.float32)

#初始化权重
w = torch.tensor(0.0,dtype=torch.float32,requires_grad=True)

# 向前传递
def forward(x):
    return w*x

# 计算损失函数
def loss(y,y_hat):
    return ((y_hat - y)**2).mean()

# mse = 1/n *(w*x - y)
# dJ/dw = 

def grad(x,y,y_hat):
    return np.dot(2*x, y_hat-y).mean()

print(f'训练前的预测值： f(5) = {forward(5):.3f}')

learning_rate = 0.05
n_iters = 20

for epoch in range(n_iters):
    y_pred = forward(x)
    # loss
    l = loss(x,y_pred)
    # grad = backward pass
    l.backward() # dl/dw
    # update weights
    with torch.no_grad():
        w -= learning_rate*w.grad
    # Zero grad
    w.grad.zero_()
    
    if epoch%1==0:
        print(f'epoch {epoch+1}:w = {w:.3f},loss = {l:.8f}')

print(f'训练后的预测值为：f(5) =  {forward(5):.3f}')




```
